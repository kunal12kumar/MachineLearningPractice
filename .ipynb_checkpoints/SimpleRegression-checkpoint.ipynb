# In this we are going to train our model with using simple Input variables


```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# importing file from the folder
salary_file=pd.read_csv('Salary_Data.csv', sep=',' , names=["expertise" , "salary"])
```

```python
salary_file.head()
```

```python
salary_file["expertise"]
```

```python
# plot the graph to see the relation bw the expeience and salary 
plt.scatter(salary_file["expertise"] ,salary_file["salary"])
plt.xlabel('Experience')
plt.ylabel('Salary')
```

```python
# now we will remove the first row shich is creating problem in the graph 
salary_file=salary_file.drop(index=0)
```

```python
# now see the relation between the experience and salary 
salary_file.corr()
```

```python
salary_file.head()
salary_file.info()
print(salary_file.columns)
```

```python
# Since the Expertise and Salary are in object therefor first convert it inot the numeric value 
salary_file['expertise']=pd.to_numeric(salary_file['expertise'], errors='coerce')
salary_file['salary']=pd.to_numeric(salary_file['salary'], errors='coerce')
```

```python
# now we can visualixe the relation between the expertise and salary 
import seaborn as sns 
sns.pairplot(salary_file)
```

# NOw training of dataset start

```python
# first classify the independent dataset
x=salary_file[['expertise']]
# as we know that only series or 2 d array can used as indepentennt variables 
print(type(x))
```

```python
# now we can change the expertise data into the array 
X=np.array(x)
X.shape
```

```python
y=salary_file['salary']
```

```python
Y=np.array(y)
Y.shape
```

```python
print(y.shape)
```

# Now split the data into test and train data

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train , y_test =  train_test_split( X, Y , test_size=0.25 , random_state=42)
```

```python
print(x_train)
```

# now standardization of data to solve the problem of the large value

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
```

```python
print(x_train)
```

```python
# now doing this for test data
x_test=scaler.transform(x_test)
```

```python
print(x_test)
```

# Now we train model by linear regression 

```python
from sklearn.linear_model import LinearRegression
simpleregression=LinearRegression()
simpleregression.fit(x_train, y_train)
```

```python
simpleregression.coef_
```

```python
simpleregression.intercept_
```

```python
plt.scatter(x_train ,y_train)
plt.scatter(x,y)
```

```python
y_predicated=simpleregression.predict(x_train)
```

```python
print(y_predicated)
```

```python
plt.plot(x_train, y_predicated)
```

# now calculate performance metrices

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_train, y_predicated)
rmse=np.sqrt(mse)
print(mse)
print(rmse)
```

```python
from sklearn.metrics import median_absolute_error
mae=median_absolute_error(y_train ,y_predicated)
print(mae)
```

# Now calculate accuracy of the model

```python
from sklearn.metrics import r2_score
rsq=r2_score(y_train , y_predicated)
```

```python
print(rsq)
```

# NOw prediction for the test data 


```python
y_testpredicted=simpleregression.predict(x_test)
print(y_testpredicted)
```

```python
ypredicted= simpleregression.intercept_  + simpleregression.coef_*(x_test)
print(ypredicted)
```

```python
plt.plot(y_testpredicted ,ypredicted)
```
